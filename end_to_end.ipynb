```python
import csv
import matplotlib.pyplot as plt
import math
from matplotlib.ticker import FuncFormatter
import numpy as np
```

```python
greedy_weights = []
shortest_weights = []
with open('../_output.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row[0] == 'greedy':
            greedy_weights.append(math.log10(float(row[2])))
        elif row[0] == 'shortest':
            shortest_weights.append(math.log10(float(row[2])))

x = range(1, len(greedy_weights) + 1)
y1 = greedy_weights
y2 = shortest_weights

plt.plot(x, y1, label="Heuristic algorithm")
plt.plot(x, y2, label="Dijkstra-based algorithm")

old_yticks = [-1,-2,-3,-4]
old_xticks = [int(x) for x in np.linspace(0, 150, 5)]

def format_y_ticks(value, pos):
    return fr"$10^{{{value}}}$"

def format_x_ticks(value, pos):
    return fr"${value}^{{th}}$"

plt.gca().yaxis.set_major_formatter(FuncFormatter(format_y_ticks))
plt.gca().set_yticks(old_yticks)
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_x_ticks))
plt.gca().set_xticks(old_xticks)

plt.xlabel('Execution')
plt.ylabel('Runtime (seconds)')
plt.xlim(left=0, right=150)
plt.legend()
plt.show()

greedy_weights = []
shortest_weights = []
with open('../_output.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row[0] == 'greedy':
            greedy_weights.append(float(row[2]))
        elif row[0] == 'shortest':
            shortest_weights.append(float(row[2]))
y1 = greedy_weights
y2 = shortest_weights
```

```python
greedy_weights = []
shortest_weights = []
with open('../_output.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row[0] == 'greedy':
            greedy_weights.append(row[2])
        elif row[0] == 'shortest':
            shortest_weights.append(row[2])

x = range(1, len(greedy_weights) + 1)
y1 = greedy_weights
y2 = shortest_weights
y_1 = np.array(y2).astype(float) / np.array(y1).astype(float)
greedy_times = []
shortest_times = []
with open('../_output.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row[0] == 'greedy':
            greedy_times.append(float(row[1]))
        elif row[0] == 'shortest':
            shortest_times.append(float(row[1]))
x = range(1, len(greedy_times) + 1)
y1 = greedy_times
y2 = shortest_times

y_2 = np.array(y2).astype(float) / np.array(y1).astype(float)
# Customization options
c1 = "lightyellow"
mean_color = "green"

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(6, 8))
axs[0].tick_params(axis='both', which='major', labelsize=14)
axs[1].tick_params(axis='both', which='major', labelsize=14)

# Box plot for the first dataset (subplot 1)
axs[0].boxplot(y_1,
               patch_artist=True,
               boxprops=dict(facecolor=c1),
               medianprops=dict(color=mean_color, linewidth=5))
axs[0].set_xticks([1])
axs[0].set_xticklabels(['Runtime ratio'])
axs[0].set_yticks(np.linspace(0,50,11))
# axs[0].set_ylim(top=20)

# Box plot for the second dataset (subplot 2)
axs[1].boxplot(y_2,
               patch_artist=True,
               boxprops=dict(facecolor=c1),
               medianprops=dict(color=mean_color, linewidth=5))
axs[1].set_xticks([1])
axs[1].set_xticklabels(['Transmission delay ratio'])

fig.suptitle(
    ' Heuristic Algorithm and the Dijkstra-based Algorithm\n Results Analysis',
    fontsize=16)

fig.tight_layout(pad=2)

plt.show()
```
